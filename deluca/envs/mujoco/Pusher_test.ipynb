```python
from deluca.envs import Pusher
import jax.numpy as np
env = Pusher()
print('action_space:' + str(env.action_space))
print('observation_space:' + str(env.observation_space)) 
x = env.get_state()
print('x:' + str(x))
u = np.ones((7,))
print('u:' + str(u))

f_x = env.f_x(x,u,0)
f_u = env.f_u(x,u,0)
f_xx = env.f_xx

print('f_x:' + str(f_x))
print('f_u:' + str(f_u))
print('f_xx:' + str(f_xx))
```
