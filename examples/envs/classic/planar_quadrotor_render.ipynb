```python
import deluca
import gym
import gnwrapper
import jax
from deluca.envs import PlanarQuadrotor
from deluca.agents import ILQR
from pyvirtualdisplay import Display
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display

def show_state(env, step=0, info=""):
    plt.figure(3)
    plt.clf()
    plt.imshow(env.render(mode='rgb_array'))
    # plt.title("%s | Step: %d %s" % (env._spec.id,step, info))
    plt.title("%s | Step: %d %s" % (type(env).__name__,step, info))
    plt.axis('off')

    display.clear_output(wait=True)
    display.display(plt.gcf())
```

```python
def loop(context, i):
    env, agent = context
    control = agent(env.state)
    _, reward, _, _ = env.step(control)
    show_state(env, step=i)
    return (env, agent), reward
```

```python
# ILQR
agent = ILQR()
agent.train(PlanarQuadrotor(wind=0), 10)
```

```python
# needed to connect to "fake" monitor (not actually used)
display_dummy = Display(visible=False, size=(1400, 900))
display_dummy.start()
```

```python
# for loop version
T = 100
env = PlanarQuadrotor(wind=0)
print(env.reset())
reward = 0
for i in range(T):
    (env, agent), r = loop((env, agent), i)
    reward += r
# env.reset()
# env.display()
reward_forloop = reward
print('reward_forloop = ' + str(reward_forloop))
```
